```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
```

```python
#Cargar el archivo datos/asaltos_yucatan.csv en un DataFrame llamado data
# si es necesario usar la opcion encoding = "ISO-8859-1"
df1 = pd.read_csv('datos/asaltos_yucatan.csv', encoding = "ISO-8859-1")
```

```python
#Mostrar las primeras filas del DataFrame data
df1.head()
```

```python
data = df1[df1['Ciudad'] == 'Mérida']
data.head()
```

```python
#Comenzando definiendo la ubicacion del mapa
map = folium.Map(location=[20.9785, -89.6332], zoom_start=11,min_zoom=8,max_zoom=15)

#Definimos un titulo para llamar posteriormente
title = 'Asaltos en Yucatán'

#Comenzamos a trabajar con los comandos en folium

#Opcion 1 de iteracion con .iterrows()#
for i, fila in data.iterrows():
    folium.Marker(location=[fila['lat'], fila['long']], popup=data['Categoria'],  
             icon=folium.Icon(color='red',icon='none') ).add_to(map)
        
#Opcion 2 de iteracion con iloc (Funciona con el df1 pero al retirar los datos de progreso y crear un nuevo dataframe#
#únicamente con los datos de mérida deja de funcionar la iteración)#
#for i in range(len(df1)):
    #folium.Marker(location=[df1.loc[i, 'lat'], df1.loc[i, 'long']], popup=df1.loc[i, 'Categoria'],  
             #icon=folium.Icon(color='red',icon='none'), legend_name=title ).add_to(map)
map
```

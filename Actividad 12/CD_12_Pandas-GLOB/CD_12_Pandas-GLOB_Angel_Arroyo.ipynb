## Concatenando todos los archivos .csv para formar un dataframe


```python
# establecer ruta de búsqueda  
# aquí queremos buscar archivos csv  
path = 'datos'
files = glob.glob(path + '/*.csv')

# crear una lista vacía para almacenar  los dataframes
lista_archivos = []

# recorrer la lista de archivos para crear la lista de archivos 
for f in files:
    # leer el archivo actual csv
    temp_df = pd.read_csv(f)
    # Adjuntar al dataframe 
    lista_archivos.append(temp_df)
    print(f' Dataframe creado con éxito  {f} con las dimensiones {temp_df.shape}')

# concatenación
df = pd.concat(lista_archivos, axis=0)
print('*Dimensiones finales',df.shape)
df.head()
```

Para realizar este ejercicio no encontré ninguna función que permita ligar un dato a su archivo de origen, por tanto decidí realizar la adición de la columna antes de unir todos los archivos. Para ello realizaré una iteración sencilla que permita agregar un mismo valor en la nueva columna llamada "nombre_archivo". Utilicé el string declarado para generar el archivo como el string que será agregado a la columna anteriormente mencionada. Definitivamente se puede automatizar el código mediante ciclos sin embargo la solución que presento es funcional.

```python
#Semestre 1a)
path = 'datos'
loc1a = glob.glob(path + '/semestre1a.csv')

lista_archivos = []
for f in loc1a:
    temp_df = pd.read_csv(f)
    lista_archivos.append(temp_df)

df1a = pd.concat(lista_archivos, axis=0)
df1a['nombre_archivo'] = pd.Series([loc1a for x in range(len(df1a.index))])
df1a.head()
```
